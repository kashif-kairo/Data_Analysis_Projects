# **Fast Delivery Agents**
Performance Analysis: Swiggy Instamart, JioMart, Zepto & Blinkit Compared

This dataset contain customer reviews and ratings for popular fast delivery agents, including Zepto, Blinkit, Swiggy Instamart, and JioMart. The dataset includes 5000 entries and covers various attributes such as the delivery agent's name, customer ratings, review text, delivery times, order types, and more. It is structured to provide insights into customer satisfaction, delivery performance, and service quality. This dataset is ideal for analysis related to customer feedback, sentiment analysis, and delivery service performance.

```python
#including improtant libaries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/zepto, swiggy, instamat.csv")
```

```python
# Overall Distribution
# how many orders each agent (Zepto, Blinkit, Swiggy Instamart, JioMart) fulfilled.

sns.countplot(x='Agent Name', data=df)
plt.title('Count of Orders by Agent')
plt.show()
agent_order_counts = df["Agent Name"].value_counts()

top_agent = agent_order_counts.idxmax()
top_orders = agent_order_counts.max()

print(f"The agent with the highest number of orders is {top_agent} with {top_orders} orders.")
```

# **Rating of each Agents**

```python
g = sns.FacetGrid(df, col="Agent Name", height=4, aspect=1)
g.map(sns.histplot, "Rating", bins=5, kde=False)
g.set_titles("{col_name}")
g.set_axis_labels("Rating", "Count")
plt.suptitle("Distribution of Ratings by Agent", y=1.05)
plt.show()

# Find the agent with the highest average rating
best_rated_agent = df.groupby("Agent Name")["Rating"].mean().idxmax()
best_avg_rating = df.groupby("Agent Name")["Rating"].mean().max()

print(f"The agent with the best average rating is {best_rated_agent} with an average rating of {best_avg_rating:.2f}.")
```

# **Observation**

*   The agent with the best average rating is Swiggy Instamart with an average rating of 3.02.


# **Customer Feedback**

```python
plt.figure(figsize=(12, 8))
positive_feedback = {}

agents = df["Agent Name"].unique()

for i, agent in enumerate(agents, 1):
    plt.subplot(2, 2, i)  # Adjust grid size if needed
    agent_feedback_counts = df[df["Agent Name"] == agent]["Customer Feedback Type"].value_counts()

    # Store positive feedback percentage
    if "Positive" in agent_feedback_counts.index:
        positive_feedback[agent] = (agent_feedback_counts["Positive"] / agent_feedback_counts.sum()) * 100
    else:
        positive_feedback[agent] = 0  # If no positive feedback exists

    plt.pie(agent_feedback_counts, labels=agent_feedback_counts.index, autopct='%1.1f%%',
            colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'], startangle=140)
    plt.title(f"Customer Feedback for {agent}")

plt.tight_layout()
plt.show()

# Find the agent with the highest positive feedback percentage
best_feedback_agent = max(positive_feedback, key=positive_feedback.get)
best_feedback_percentage = positive_feedback[best_feedback_agent]

print(f"The agent with the highest positive customer feedback is {best_feedback_agent} with {best_feedback_percentage:.2f}% positive feedback.")
```



*   Swiggy Instamart has the highest positive feedback among all agents, indicating better customer satisfaction compared to Zepto, Blinkit, and JioMart.
*   Zepto and Blinkit show a more balanced mix of positive and negative feedback, suggesting room for improvement in service quality.
*   JioMart appears to have the lowest positive feedback percentage, which might indicate issues with service, delivery, or customer experience.

# **Discount Analysis**

```python
discount_counts = df[df["Discount Applied"] != "No"].groupby("Agent Name")["Discount Applied"].count()

# Plot the results
plt.figure(figsize=(8, 5))
sns.barplot(x=discount_counts.index, y=discount_counts.values, palette="viridis")
plt.title("Total Discounts Given by Each Agent")
plt.xlabel("Agent Name")
plt.ylabel("Number of Discounts Applied")
plt.xticks(rotation=45)
plt.show()

# Find the agent with the highest discounts
most_discounts_agent = discount_counts.idxmax()
print(f"The agent that provides the most discounts is {most_discounts_agent}.")
```

# **Delivery Performance Analysis**

```python
# 3. Average Delivery Time per Agent
avg_delivery_time = df.groupby("Agent Name")["Delivery Time (min)"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=avg_delivery_time.index, y=avg_delivery_time.values, palette='magma')
plt.title('Average Delivery Time by Agent')
plt.xlabel('Agent Name')
plt.ylabel('Average Delivery Time (mins)')
plt.xticks(rotation=45)
plt.show()

# 4. On-Time Delivery Percentage
on_time_delivery = df[df["Delivery Time (min)"] <= 30].groupby("Agent Name")["Delivery Time (min)"].count() / df.groupby("Agent Name")["Delivery Time (min)"].count() * 100

plt.figure(figsize=(8, 5))
sns.barplot(x=on_time_delivery.index, y=on_time_delivery.values, palette='Blues')
plt.title('On-Time Delivery Percentage by Agent')
plt.xlabel('Agent Name')
plt.ylabel('On-Time Delivery (%)')
plt.xticks(rotation=45)
plt.show()
```

# **Customer Satisfaction & Service Analysis**

```python
# 5. Average Ratings by Agent
avg_ratings = df.groupby("Agent Name")["Rating"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=avg_ratings.index, y=avg_ratings.values, palette='coolwarm')
plt.title('Average Customer Ratings by Agent')
plt.xlabel('Agent Name')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.show()

# 6. Customer Service Rating
avg_customer_service = df.groupby("Agent Name")["Customer Service Rating"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=avg_customer_service.index, y=avg_customer_service.values, palette='plasma')
plt.title('Customer Service Rating by Agent')
plt.xlabel('Agent Name')
plt.ylabel('Customer Service Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Discounts Given by Each Agent
discount_counts = df[df["Discount Applied"] != "No"].groupby("Agent Name")["Discount Applied"].count()

plt.figure(figsize=(8, 5))
sns.barplot(x=discount_counts.index, y=discount_counts.values, palette='Set2')
plt.title('Total Discounts Given by Each Agent')
plt.xlabel('Agent Name')
plt.ylabel('Number of Discounts Applied')
plt.xticks(rotation=45)
plt.show()

# 8. Price Range Distribution
plt.figure(figsize=(10, 5))
sns.countplot(x='Price Range', hue='Agent Name', data=df, palette='husl')
plt.title('Price Range Distribution by Agent')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title="Agent Name")
plt.show()
```

```python
# 9. Product Availability Score by Agent
product_availability_counts = df.groupby("Agent Name")["Product Availability"].value_counts(normalize=True).unstack() * 100

plt.figure(figsize=(10, 5))
product_availability_counts.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 5))
plt.title('Product Availability Percentage by Agent')
plt.xlabel('Agent Name')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title="Availability Status")
plt.show()

```

```python
# Grouping by Location and Agent Name to find key statistics
location_performance = df.groupby(['Location', 'Agent Name']).agg({
    'Rating': 'mean',  # Average rating per agent in each location
    'Delivery Time (min)': 'mean',  # Average delivery time per agent
    'Customer Service Rating': 'mean',  # Average customer service rating
    'Agent Name': 'count'  # Total orders handled
}).rename(columns={'Agent Name': 'Total Orders'}).reset_index()

# Finding the best agent in each location based on Ratings
best_agents_by_location = location_performance.loc[location_performance.groupby("Location")["Rating"].idxmax()]

# Display the best agents for each location
print(best_agents_by_location)

# Visualization - Best Rated Agent per Location
plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Rating', hue='Agent Name', data=best_agents_by_location, palette='coolwarm')
plt.title('Best Rated Agent in Each Location')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.legend(title="Agent Name")
plt.show()
```

```python
# Finding the best feature of each agent

best_features = {}

# Fastest Delivery
fastest_agent = df.groupby("Agent Name")["Delivery Time (min)"].mean().idxmin()
best_features[fastest_agent] = "Fastest Delivery"

# Highest Order Accuracy
most_accurate_agent = df.groupby("Agent Name")["Order Accuracy"].apply(lambda x: (x == "Accurate").mean()).idxmax()
best_features[most_accurate_agent] = "Most Accurate Orders"

# Best Customer Ratings
highest_rated_agent = df.groupby("Agent Name")["Rating"].mean().idxmax()
best_features[highest_rated_agent] = "Best Customer Ratings"

# Most Discounts Given
most_discount_agent = df.groupby("Agent Name")["Discount Applied"].apply(lambda x: (x == "Yes").sum()).idxmax()
best_features[most_discount_agent] = "Most Discounts Offered"

# Print Best Features of Each Agent
for agent, feature in best_features.items():
    print(f" {agent} - Best Feature: {feature}")
```

# ** Insights & Summary:**
1 Swiggy Instamart & Blinkit might fulfill the most orders. 

2 Zepto & Blinkit might have the fastest delivery times. 

3 JioMart might give the most discounts, while Swiggy Instamart could have higher customer service ratings. 

4 Most positive customer feedback could belong to Blinkit or Zepto based on sentiment analysis.

5 Price-sensitive customers may prefer JioMart or Blinkit, depending on the price range data.

```python
# Assigning points (higher is better for rating/feedback, lower is better for delivery time)
df["Rating Rank"] = df.groupby("Agent Name")["Rating"].rank(ascending=False)
df["Delivery Rank"] = df.groupby("Agent Name")["Delivery Time (min)"].rank(ascending=True)
df["Feedback Rank"] = df.groupby("Agent Name")["Customer Feedback Type"].transform(lambda x: (x == "Positive").sum())
df["Discount Rank"] = df.groupby("Agent Name")["Discount Applied"].transform(lambda x: (x == "Yes").sum())

# Aggregating points per agent
agent_scores = df.groupby("Agent Name")[["Rating Rank", "Delivery Rank", "Feedback Rank", "Discount Rank"]].mean()
agent_scores["Total Score"] = agent_scores.sum(axis=1)

# Finding the overall winner
overall_winner = agent_scores["Total Score"].idxmax()
print(f" The overall best agent is: {overall_winner}")
```
